```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = {"id":[], "timestamp":[], "latitude":[], "longitude":[], "altitude":[], "UTM_easting":[], "UTM_northing":[], "zone_number":[], "zone_code":[]}
```

```python
filepath = '/home/ubuntu20_04/EECE5554/LAB1/src/bag_files/covered_location.csv'
with open(filepath, "r") as f:
    next(f)
    lines = csv.reader(f, delimiter=",")
    for i,row in enumerate(lines):
        data['id'].append(i)
        data['timestamp'].append(row[0])
        data['latitude'].append(row[7])
        data['longitude'].append(row[8])
        data['altitude'].append(row[9])
        data['UTM_easting'].append(row[11])
        data['UTM_northing'].append(row[12])
        data['zone_number'].append(row[14])
        data['zone_code'].append(row[15])
data["timestamp"] = np.asarray(data["timestamp"])
data["latitude"] = np.asarray(data["latitude"], dtype = 'float64')
data["longitude"] = np.asarray(data["longitude"], dtype = 'float64')
data["altitude"] = np.asarray(data["altitude"], dtype = 'float64')
data["UTM_easting"] = np.asarray(data["UTM_easting"], dtype = 'float64')
data["UTM_northing"] = np.asarray(data["UTM_northing"], dtype = 'float64')
data["zone_number"] = np.asarray(data["zone_number"], dtype = 'int32')
data["zone_code"] = np.asarray(data["zone_code"])
```

```python
data_open = {"id":[], "timestamp":[], "latitude":[], "longitude":[], "altitude":[], "UTM_easting":[], "UTM_northing":[], "zone_number":[], "zone_code":[]}
```

```python
filepath = '/home/ubuntu20_04/EECE5554/LAB1/src/bag_files/open_location.csv'
with open(filepath, "r") as f:
    next(f)
    lines = csv.reader(f, delimiter=",")
    for i,row in enumerate(lines):
        data_open['id'].append(i)
        data_open['timestamp'].append(row[0])
        data_open['latitude'].append(row[7])
        data_open['longitude'].append(row[8])
        data_open['altitude'].append(row[9])
        data_open['UTM_easting'].append(row[11])
        data_open['UTM_northing'].append(row[12])
        data_open['zone_number'].append(row[14])
        data_open['zone_code'].append(row[15])
data_open["timestamp"] = np.asarray(data_open["timestamp"])
data_open["latitude"] = np.asarray(data_open["latitude"], dtype = 'float64')
data_open["longitude"] = np.asarray(data_open["longitude"], dtype = 'float64')
data_open["altitude"] = np.asarray(data_open["altitude"], dtype = 'float64')
data_open["UTM_easting"] = np.asarray(data_open["UTM_easting"], dtype = 'float64')
data_open["UTM_northing"] = np.asarray(data_open["UTM_northing"], dtype = 'float64')
data_open["zone_number"] = np.asarray(data_open["zone_number"], dtype = 'int32')
data_open["zone_code"] = np.asarray(data_open["zone_code"])
```

```python
start_point = [data["UTM_easting"][0], data["UTM_northing"][0]]
end_point = [data["UTM_easting"][-1], data["UTM_northing"][-1]]

start_point_open = [data_open["UTM_easting"][0], data_open["UTM_northing"][0]]
end_point_open = [data_open["UTM_easting"][-1], data_open["UTM_northing"][-1]]
```

```python
fig, (ax1,ax2) = plt.subplots(2)
fig.tight_layout(pad=5.0)
ax1.plot(data["UTM_easting"],data["UTM_northing"])
ax2.plot(data_open["UTM_easting"],data_open["UTM_northing"])
ax1.scatter(start_point[0], start_point[1], label="Start Point")
ax1.scatter(end_point[0], end_point[1], label="End Point")
ax2.scatter(start_point_open[0], start_point_open[1], label="Start Point Open")
ax2.scatter(end_point_open[0], end_point_open[1], label="End Point Open")

ax1.set_title("Occluded Area")
ax2.set_title("Open Area")
ax1.set_xlabel("Easting (metres)")
ax1.set_ylabel("Northing (metres)")
ax2.set_xlabel("Easting (metres)")
ax2.set_ylabel("Northing (metres)")
ax1.grid(which='major', color='#AAAAAA', linewidth=1)
ax1.grid(which='minor', color='#AAAAAA', linestyle=':', linewidth=1)
ax2.grid(which='major', color='#AAAAAA', linewidth=1)
ax2.grid(which='minor', color='#AAAAAA', linestyle=':', linewidth=1)
ax1.minorticks_on()
ax2.minorticks_on()
ax1.grid(True)
ax2.grid(True)
ax1.legend()
ax2.legend()
plt.show()
```

```python
data_walk = {"id":[], "timestamp":[], "latitude":[], "longitude":[], "altitude":[], "UTM_easting":[], "UTM_northing":[], "zone_number":[], "zone_code":[]}

filepath = '/home/ubuntu20_04/EECE5554/LAB1/src/bag_files/straight_line.csv'
with open(filepath, "r") as f:
    next(f)
    lines = csv.reader(f, delimiter=",")
    for i,row in enumerate(lines):
        data_walk['id'].append(i)
        data_walk['timestamp'].append(row[0])
        data_walk['latitude'].append(row[7])
        data_walk['longitude'].append(row[8])
        data_walk['altitude'].append(row[9])
        data_walk['UTM_easting'].append(row[11])
        data_walk['UTM_northing'].append(row[12])
        data_walk['zone_number'].append(row[14])
        data_walk['zone_code'].append(row[15])
data_walk["timestamp"] = np.asarray(data_walk["timestamp"])
data_walk["latitude"] = np.asarray(data_walk["latitude"], dtype = 'float64')
data_walk["longitude"] = np.asarray(data_walk["longitude"], dtype = 'float64')
data_walk["altitude"] = np.asarray(data_walk["altitude"], dtype = 'float64')
data_walk["UTM_easting"] = np.asarray(data_walk["UTM_easting"], dtype = 'float64')
data_walk["UTM_northing"] = np.asarray(data_walk["UTM_northing"], dtype = 'float64')
data_walk["zone_number"] = np.asarray(data_walk["zone_number"], dtype = 'int32')
data_walk["zone_code"] = np.asarray(data_walk["zone_code"])
start_point = [data_walk["UTM_easting"][0], data_walk["UTM_northing"][0]]
end_point = [data_walk["UTM_easting"][-1], data_walk["UTM_northing"][-1]]
start_point_alt = [data_walk["id"][0], data_walk["altitude"][0]]
end_point_alt = [data_walk["id"][-1], data_walk["altitude"][-1]]
```

```python
fig, ax = plt.subplots()
ax.plot(data_walk["UTM_easting"],data_walk["UTM_northing"])
ax.scatter(start_point[0], start_point[1], label="Start Point")
ax.scatter(end_point[0], end_point[1], label="End Point")

ax.set_title("Walking data")
ax.set_xlabel("Easting (metres)")
ax.set_ylabel("Northing (metres)")
ax.grid(which='major', color='#AAAAAA', linewidth=1)
ax.grid(which='minor', color='#AAAAAA', linestyle=':', linewidth=1)
ax.minorticks_on()

ax.grid(True)
ax.legend()
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(data_walk["id"],data_walk["altitude"])
ax.scatter(start_point_alt[0], start_point_alt[1], label="Start Point")
ax.scatter(end_point_alt[0], end_point_alt[1], label="End Point")

ax.set_title("Walking data")
ax.set_xlabel("Timesteps (seconds)")
ax.set_ylabel("Altitude (meteres)")
ax.grid(which='major', color='#AAAAAA', linewidth=1)
ax.grid(which='minor', color='#AAAAAA', linestyle=':', linewidth=1)

ax.minorticks_on()
ax.grid(True)
ax.legend()
plt.show()
```

```python
m_walk, c_walk = np.polyfit(data_walk["UTM_easting"], data_walk["UTM_northing"], 1)
yn_walk = np.polyval([m_walk, c_walk], data_walk["UTM_easting"])
```

```python
fig, ax = plt.subplots()
ax.plot(data_walk["UTM_easting"], data_walk["UTM_northing"])
ax.scatter(start_point[0], start_point[1], label="Start Point")
ax.scatter(end_point[0], end_point[1], label="End Point")

print(np.max(data_walk["UTM_northing"]-yn_walk))
print(np.min(data_walk["UTM_northing"]-yn_walk))
ax.plot(data_walk["UTM_easting"],yn_walk,'--', label="Best fit line", color='g')
error = np.sum(data_walk["UTM_northing"]-yn_walk)
print(error)
# ax.tight_layout()
ax.set_title("straight line")
ax.set_xlabel("Easting (metres)")
ax.set_ylabel("Northing (metres)")
ax.grid(which='major', color='#AAAAAA', linewidth=1)
ax.grid(which='minor', color='#AAAAAA', linestyle=':', linewidth=1)
ax.minorticks_on()
ax.grid(True)
ax.legend()
plt.show()
```

```python
known_occ_lat = 42.3375865
known_occ_lon = -71.0902112
known_op_lat = 42.337138
known_op_lon = -71.089898
err_oc_lat = data['latitude'] - known_occ_lat
err_oc_lon = data['longitude'] - known_occ_lon
err_op_lat = data_open['latitude'] - known_op_lat
err_op_lan = data_open['longitude'] - known_op_lon
```

```python
mean_oc_lat = np.mean(err_oc_lat)
mean_op_lat = np.mean(err_op_lat)
mean_oc_lon = np.mean(err_oc_lon)
mean_op_lan = np.mean(err_op_lan)
```

```python
print("Mean of error for latitude for occluded space:", mean_oc_lat)
print("Mean of error for longitude for occluded space:", mean_oc_lon)
print("Mean of error for latitude for open space:", mean_op_lat)
print("Mean of error for latitude for open space:", mean_op_lan)
```

```python
median_oc_lat = np.median(err_oc_lat)
median_op_lat = np.median(err_op_lat)
median_oc_lon = np.median(err_oc_lon)
median_op_lan = np.median(err_op_lan)
```

```python
print("Median of error for latitude for occluded space:", median_oc_lat)
print("Median of error for longitude for occluded space:", median_oc_lon)
print("Median of error for latitude for open space:", median_op_lat)
print("Median of error for latitude for open space:", median_op_lan)
```
