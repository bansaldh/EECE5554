```python
import csv
import matplotlib.pyplot as plt
import numpy as np
```

```python
def load_data(filepath):
    data = {"name": "not_set", "id":[], ".Header.frame_id":[], ".Header.stamp.secs":[], ".Latitude":[], ".Longitude":[], "Altitude":[], '.fixed_quality':[], "easting":[], 
	 "northing":[], ".Letter":[], ".Zone":[], ".HDOP":[]}
    with open(filepath, "r") as file:
        next(file)
        lines = csv.reader(file, delimiter = ",")
        for i, row in enumerate(lines):
            data["id"].append(i)
            data[".Header.stamp.secs"].append(row[2])
            data[".Latitude"].append(float(row[5]))
            data[".Longitude"].append(float(row[6]))
            data["Altitude"].append(float(row[7]))
            data[".HDOP"].append(float(row[8]))
            data[".fixed_quality"].append(float(row[14]))
            data["easting"].append(float(row[9]))
            data["northing"].append(float(row[10]))
            data[".Letter"].append(row[13])
            data[".Zone"].append(int(row[12]))

        data["id"] = np.asarray(data["id"])
        data[".Latitude"] = np.asarray(data[".Latitude"], dtype = 'float64')
        data[".Longitude"] = np.asarray(data[".Longitude"], dtype = 'float64')
        data["Altitude"] = np.asarray(data["Altitude"], dtype = 'float64')
        data[".fixed_quality"] = np.asarray(data[".fixed_quality"], dtype = 'float64')
        data["easting"] = np.asarray(data["easting"], dtype = 'float64')
        data["northing"] = np.asarray(data["northing"], dtype = 'float64')
        data[".Letter"] = np.asarray(data[".Letter"])
        data[".Zone"] = np.asarray(data[".Zone"], dtype = 'int32')
        data[".HDOP"] = np.asarray(data[".HDOP"], dtype = 'int32')

    return data
```

```python
stat_open = load_data("/home/ubuntu20_04/EECE5554/LAB2/src/Analysis/csv_files/stationary_open.csv")
stat_occ = load_data("/home/ubuntu20_04/EECE5554/LAB2/src/Analysis/csv_files/stationary_par_occluded.csv")
move_open = load_data("/home/ubuntu20_04/EECE5554/LAB2/src/Analysis/csv_files/moving_open.csv")
move_occ = load_data("/home/ubuntu20_04/EECE5554/LAB2/src/Analysis/csv_files/moving_par_occluded.csv")
```

```python
X = stat_open["easting"]
Y = stat_open["northing"]
name = stat_open["name"]
Start_point = [X[0], Y[0]]
End_point = [X[-1], Y[-1]]

fig = plt.figure()
ax = fig.add_subplot()
ax.grid(which='major', color='#AAAAAA', linewidth=1)
ax.grid(which='minor', color='#AAAAAA', linestyle=':', linewidth=1)
ax.minorticks_on()
ax.grid(True)

ax.scatter(X,Y, marker='+', s=50, label="Collected data points")
ax.scatter(Start_point[0], Start_point[1], label="Start Point")
ax.scatter(End_point[0], End_point[1], label="End Point")


ax.set_title("stationary open" +" plot")
ax.set_xlabel("Easting (meters)")
ax.set_ylabel("Northing (meters)")

ax.legend()
plt.show()
```
